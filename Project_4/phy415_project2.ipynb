## 0. Attractive potential between two objects & model assumptions

#### Here, we are investigating the attractive potential between 2 moving objects using: 

$$V(r) = \frac{-\alpha}{r}$$

The goal of this project is to better understand the types of motion that this type of potential can bring about, notably analyzing these possible modes of motion through kinetic and potential energy balances. It is also interesting to look at from an astronomy perspective. Binary star systems can operate in many ways, and typically they fall into stable orbits, where they revolve around a center of mass that may or may not move. Here we will also look at some of the conditions required for certain types of orbits, try to categorize and differentiate them, and also look at the stability of certain systems.


What is a stable orbit?

What is an unbounded/bounded orbit?

Is there a difference between a bounded orbit and a stable orbit?

Can an unstable system eventually stabilize? If so, why is this able to happen?


These are all the questions I am looking to answer. To do it, I will be analytically and numerically analyzing this potential looking at the center of mass of the system, energy ratios, energy conservations, as well as positions and velocities.

#### Model assumptions:

This is a very simple model. Since I want to mainly focus on understanding the basics of orbit stabilities, I introduce these conditions:

1) both masses are the same, and are equal to 1kg.

## 1. Lagrangian Examination

#### Energy considerations:

There are two energies that contribute to the overall energy of this system. Kinetic energy, and potential energy.

The kinetic energy arises from the velocity of the two moving objects. We can write:

$$K = \frac{m}{2}(\dot{r_{1}}^2 + \dot{r_{2}}^2) $$

As we talked about earlier, there is one energy contribution from the potential energy. We can also write this as:

$$U = \frac{-\alpha}{|r_2 - r_1|} $$

The resulting Lagrangian is $$ L = K-U = \frac{m}{2}(\dot{r_{1}}^2 + \dot{r_{2}}^2)+\frac{\alpha}{|r_2 - r_1|}$$

-----------------------------------------------------------------------------------------------------------------------------------
#### Deriving the equations of motion for each object:

The way this is done is by taking derivatives within the lagrangian.

The equation for the first and second objects respectively will be found using 

$$\frac{\partial{L}}{\partial{r_1}} = \frac{d}{dt}\frac{\partial{L}}{\partial{\dot{r_1}}}$$

$$\frac{\partial{L}}{\partial{r_2}} = \frac{d}{dt}\frac{\partial{L}}{\partial{\dot{r_2}}}$$

For the left side if the above equations, we find 

$$\frac{\partial}{\partial{r_1}}\left(\frac{\alpha}{|r_2 - r_1|^2}\right) = \frac{-\alpha}{r}\hat{r_1}$$ 

$$\frac{\partial}{\partial{r_2}}\left(\frac{\alpha}{|r_2 - r_1|^2}\right) = \frac{\alpha}{r}\hat{r_2}$$ 

And for the right side:

$$\frac{d}{dt}\frac{\partial}{\partial{r_1}}\left(\frac{m}{2}(\dot{r_{1}}^2 + \dot{r_{2}}^2)\right) = \frac{d}{dt}m\dot{r_1} = m\ddot{r_1}$$

$$\frac{d}{dt}\frac{\partial}{\partial{r_2}}\left(\frac{m}{2}(\dot{r_{1}}^2 + \dot{r_{2}}^2)\right) = \frac{d}{dt}m\dot{r_2} = m\ddot{r_2}$$

Now, we can set these equal to each other and obtain the force equations.

$$ m\ddot{r_1} = \frac{-\alpha}{r^2}\hat{r_1}$$

$$ m\ddot{r_2} = \frac{\alpha}{r^2}\hat{r_2} $$

We can notice these results are consistent with newton's third law, as we would expect from this potential, where $F_{21} = -F_{12}$

#### Discussion of energy balances, and energy conservation

Here, you will see different types of potential graphs with the same starting potential. Look at this from a standpoint of energy conservation. Since there are no external forces on the system, energy in this system is conserved.

---------------------------------------------------------------------------------------------------------------------------
#### Types of resulting motion

The total energy is defined by the initial conditions of the system, and the radius moved constrained between the maximum and minimum values of the potential.

In the first graph, the total energy is greater than the initial potential energy, but is still overall negative. The distance between the objects will stay within the range of the starting potential, and the total energy. Here, since the total energy is negative, the distance between the point will oscillate back and forth between the two energies. This is a bounded orbit. The reason the distance $r$ cannot cross these boundaries is because of the initial kinetic and potential energies of the system. If you were to go below the point of initial potential, that would mean that energy would need to be taken away in the form of adding negative kinetic, which is not possible because kinetic energy is always positive. If you were you go above the point of total energy, you would need to add kinetic energy. This is not possible in this scenario because energy is conserved here so no energy can be added. The motion of the two object stays within the bounds of the initial energies of the system. _This is the case of a bounded orbit_.

In the second, the total energy is greater than the total energy and positive. We can see that since the total energy is positive, as $r$ increases, there is no force driving the two objects together, and the objects drift forever. The upper limit of the potential bound is 0. _This is the case of an unbounded orbit_.

In the third graph, we see that the total energy is negative and less than the potential. I just wanted to showcase this because this type of energy balance does not correlate with real motion, since the kinetic element would have to be negative. We know once again that negative kinetic energy is not possible.

_The main takeaway here is that a condition for the bounded orbit is $|V|>K$_

## 3. Center of mass and initial conditions

#### Short talk about two different types of motion arising from initial conditions

Lets consider a center of mass: $$R = \frac{m_1r_1+m_2r_2}{M}$$ where $M = m_1+m_2=2$ 

If we set $R=0$, since $m_1 = m_2$, we can see $r_1 = -r_2$. This is a possible initial condition for a bounded orbit. 
In this case $\dot{R} = 0$ and $\dot{r_1} = -\dot{r_2}$, and the center of mass will not move. Both masses will revolve quite nicely about the center.

If we have $R = constant$, we find $\dot{R} = constant$ as well, and so in this case we will see the center of mass moving as the two masses move. I will be liberal with my choice of initial conditions here, because any choice will result in a constant value for the center of mass and its velocity so I won't mathematically include any restrictions.

#### Methods

I will be using the velocity verlet method as my integration method, It is what I found to be easier for a coupled, 2nd order ODE in 4 dimensions. It is also something I found to be an accurate integrator for a past project which was also having to do with another sort of attractive potential. So I know this method works well with systems of comparable sensilbility but I will prove this by showing the error rate when plotting energy conservation.

I want to look into each motion by plotting their position in the x/y plane, and show their relative positions over time and compare that to the potential energy graph with the bounds of the total energy of the system to show that the radius will oscillate within the bounds of minimum and maximum potential.

I want to note that any minor inconsistencies in the graphs, notably in the energy graphs and in cases with larger $\alpha$ values, that these are totally normal and due to the slight errors resulting from the velocity-verlet method.

```python
def get_vel_acc(x2, x1, vx2, vx1, alpha, obj=0, m=1):
    
    '''
    given:
    position x1, x2, velocity v1, v2, 
    coefficient alpha,
    object number specification: (1 if x1, 2 if x2)
    
    returns:
    velocity and acceleration for either x1 or x2.
    
    each of these are 2d arrays with an x and y component.
    
    NOTE X2 and V2 COME BEFORE X1 and V1
    
    '''
    
    if obj == 1:
        sign = -1
        v = vx1
        
    elif obj == 2:
        sign = +1
        v = vx2
    
    rel = x2 - x1
    rabs = (rel[0]**2 + rel[1]**2)**0.5
    acc = -(alpha/rabs**3)*rel*sign/m
    vel = v
    
    if obj == 0:
        print("specify object argument. if getting acc for x1, enter 1.")
    
    return [vel, acc]
    
```

```python
def velocity_verlet(n, dt, alpha, x2_0, x1_0, v2_0, v1_0):
    
    '''
    given:
    number of points n
    timestep dt
    value alpha
    position x1, x2, velocity v1, v2, 
    coefficient alpha,
    
    returns:
    time array and position arrays for x2, x1 and velocity arrays v2, v1
    
    each of these are 2d arrays with an x and y component.
    '''
    
    x2 = [x2_0]
    x1 = [x1_0]
    v2 = [v2_0]
    v1 = [v1_0]
    
    t = np.linspace(0, n*dt, n)

    for i in range(n-1):
        
        #getting acceleration at this point
        vel2, a2 = get_vel_acc(x2[i], x1[i], v2[i], v1[i], alpha, obj = 2)
        vel1, a1 = get_vel_acc(x2[i], x1[i], v2[i], v1[i], alpha, obj = 1)
        
        #updating position
        new_x2 = x2[i] + v2[i]*dt + 0.5*a2*dt*dt
        new_x1 = x1[i] + v1[i]*dt + 0.5*a1*dt*dt
        
        #appending new positions
        x2.append(new_x2)
        x1.append(new_x1)
        
        #obtaining new acceleration
        vel2, new_a2 = get_vel_acc(new_x2, new_x1, v2[i], v1[i], alpha, obj = 2)
        vel1, new_a1 = get_vel_acc(new_x2, new_x1, v2[i], v1[i], alpha, obj = 1)
        
        #updating new velocities
        new_v2 = v2[i] + 0.5*dt*(a2 + new_a2)
        new_v1 = v1[i] + 0.5*dt*(a1 + new_a1)
        
        #appending new velocities
        v2.append(new_v2)
        v1.append(new_v1)
        
    return t, np.array(x2), np.array(x1), np.array(v2), np.array(v1)    
```

```python
def get_kin_pot(x2, x1, v2, v1, alpha, m=1):
    
    '''
    given:
    position x1, x2, velocity v1, v2, 
    coefficient alpha
    
    returns:
    kinetic, potential, and total energies of the system.
    
    each of these are 1d arrays.
    '''
    
    v1abs = (v1[:,0]**2 + v2[:,1]**2)**0.5
    v2abs = (v1[:,0]**2 + v2[:,1]**2)**0.5
    rrel = x2 - x1
    rabs = (rrel[:,0]**2 + rrel[:,1]**2)**0.5
    
    K = 0.5*m*(v1abs**2 + v2abs**2)
    U = -alpha/rabs
    E = K + U
    
    return K,U,E
```

## 4.1 - bounded case #1 with a low alpha value

```python
#initializing the system and getting positions, velocities, and energies

n = 100000
dt = 0.01
alpha = 9.1
m = 1

x20_1 = np.array([-0.5,0])
x10_1 = np.array([0.5,0])
v20_1 = np.array([0, -3])
v10_1 = np.array([0,3])

t, x2, x1, v2, v1 = velocity_verlet(n, dt, alpha, x20_1, x10_1, v20_1, v10_1)
K, U, E = get_kin_pot(x2, x1, v2, v1, alpha)
```
